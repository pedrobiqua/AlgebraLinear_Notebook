## Igualdade de vetores
Esse conceito é bastante simples para um vetor ser igual a outro ele deve:
- O $x$ do primeiro vetor for igual ao $x$ do segundo vetor
- O $y$ do primeiro vetor for igual ao $y$ do segundo vetor
- O $z$ do primeiro vetor for igual ao $z$ do segundo vetor ($r^3$)
- O $w$ do primeiro vetor for igual ao $w$ do segundo vetor ($r^4$)

```python
# Criar dois vetores iguais
vetor_a = np.array([1, 2, 3])
vetor_b = np.array([1, 2, 3])

# Verificar se os vetores são iguais
sao_iguais = np.array_equal(vetor_a, vetor_b)

# Imprimir os vetores e o resultado da comparação
print("Vetor A:", vetor_a)
print("Vetor B:", vetor_b)
print("Os vetores são iguais?", sao_iguais)
```

##Subtração de 2 Vetores

A subtração de dois vetores é realizada componente por componente, onde a componente $i$-ésima do vetor resultante é a diferença das componentes correspondentes dos dois vetores originais.

Se tivermos dois vetores $\vec{A} = \langle A_1, A_2, \ldots, A_n \rangle$ e $\vec{B} = \langle B_1, B_2, \ldots, B_n \rangle$, então a subtração $\vec{C} = \vec{A} - \vec{B}$ é dada por:

$$C_i = A_i - B_i$$

O vetor resultante $\vec{C}$ terá as mesmas dimensões que $\vec{A}$ e $\vec{B}$, e suas componentes serão as diferenças das componentes correspondentes dos vetores originais.

Em Python usando NumPy, a subtração de vetores pode ser realizada diretamente, como mostrado no exemplo:


```python
# Criar dois vetores
vetor_a = np.array([1, 2, 3])
vetor_b = np.array([4, 5, 6])

# Subtrair os vetores
vetor_resultante = vetor_a - vetor_b

# Imprimir o resultado
print("Vetor A:", vetor_a)
print("Vetor B:", vetor_b)
print("Resultado (Vetor A - Vetor B):", vetor_resultante)
```

```python
# Criar dois vetores
vetor_a = np.array([1, 2, 3])
vetor_b = np.array([4, 5, 6])

# Subtrair os vetores
vetor_resultante = vetor_a - vetor_b

# Plotar os vetores
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

# Adicionar os vetores ao plot
ax.quiver(0, 0, vetor_a[0], vetor_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vetor A')
ax.quiver(0, 0, vetor_b[0], vetor_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vetor B')
ax.quiver(0, 0, vetor_resultante[0], vetor_resultante[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vetor Resultante')

# Configurar o gráfico
ax.set_xlim(-4, 7)
ax.set_ylim(-4,7)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax.legend()

# Mostrar o plot
plt.title("Subtração de Vetores e Plots")
plt.show()
```

## Soma de 2 Vetores

A soma de dois vetores $\vec{A} = \langle A_x, A_y \rangle$ e $\vec{B} = \langle B_x, B_y \rangle$ é realizada componente por componente. O vetor resultante $\vec{C} = \vec{A} + \vec{B}$ é dado por:

$ C_x = A_x + B_x $
$ C_y = A_y + B_y $

Em Python, usando NumPy, a soma de vetores pode ser realizada de forma eficiente. Por exemplo:

```python
# Definir os vetores
A = np.array([2, 3])
B = np.array([4, 1])

# Calcular a soma dos vetores
C = A + B

fig, ax = plt.subplots(1, 1, figsize=(8, 8))

# Adicionar os vetores ao plot
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vetor A')
# Resultado da operação B - A
ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vetor B')
# Adicionar os vetores ao plot
ax.quiver(0, 0, C[0], C[1], angles='xy', scale_units='xy', scale=1, color='m', label='Vetor C')

# Configurar o gráfico
ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax.legend()

# Mostrar o plot
plt.title("Vetor AB e Vetor B - A")
plt.show()

# Resultado: C = [6, 4]
```

## Produto Escalar (Mutiplicação de vetores)

## Produto Escalar de 2 Vetores

O produto escalar (ou produto ponto) de dois vetores $\vec{A} = \langle A_1, A_2, \ldots, A_n \rangle$ e $\vec{B} = \langle B_1, B_2, \ldots, B_n \rangle$ é calculado multiplicando as componentes correspondentes e somando o resultado:

$\vec{A} \cdot \vec{B} = A_1 \cdot B_1 + A_2 \cdot B_2 + \ldots + A_n \cdot B_n$

Em Python usando NumPy, o produto escalar pode ser calculado diretamente, como mostrado no exemplo:


```python
# Definir os vetores
A = np.array([2, 3])
B = np.array([4, 1])

# Calcular o produto escalar
produto_escalar = np.dot(A, B)
print("A * B = ", produto_escalar)
```

## Multiplicação por um escalar

A multiplicação de um vetor $\vec{A} = \langle A_x, A_y \rangle$ por um escalar $k$ é realizada multiplicando cada componente do vetor pelo escalar:

$ k \cdot \vec{A} = \langle k \cdot A_x, k \cdot A_y \rangle $

Em Python usando NumPy, a multiplicação por um escalar pode ser realizada de maneira direta, como exemplificado abaixo:

```python
# Definir o vetor
A = np.array([2, 3])

# Multiplicar o vetor por um escalar
k = 3
resultado = k * A
print(resultado)

# Resultado: resultado = [6, 9]
```

## Módulo de um Vetor

O módulo (ou magnitude) de um vetor $\vec{V} = \langle V_x, V_y \rangle$ é calculado usando a fórmula euclidiana:

$\|\vec{V}\| = \sqrt{V_x^2 + V_y^2}$

Em Python, usando NumPy, o módulo de um vetor pode ser calculado de forma direta, como mostrado no exemplo:

```python
# Definir o vetor
V = np.array([1, 4, 5])

# Calcular o módulo do vetor
modulo = np.linalg.norm(V)
print("Módulo = ", modulo)
```

## Ângulo entre dois vetores(ângulo teta entre vetores)

O ângulo entre dois vetores $\vec{A}$ e $\vec{B}$ é uma medida angular que indica o quão próximos ou distantes esses vetores estão um do outro. O cálculo do ângulo entre dois vetores pode ser realizado utilizando o produto escalar e a fórmula trigonométrica do cosseno. Se $\theta$ representa o ângulo entre os vetores, então:

$$ \cos(\theta) = \frac{\vec{A} \cdot \vec{B}}{\|\vec{A}\| \|\vec{B}\|} $$

onde $\vec{A} \cdot \vec{B}$ é o produto escalar dos vetores $\vec{A}$ e $\vec{B}$, e $\|\vec{A}\|$ e $\|\vec{B}\|$ são seus módulos (ou magnitudes).

Em Python, usando NumPy, o cálculo do ângulo entre dois vetores pode ser realizado de forma eficiente, como mostrado no exemplo:


```python
# Definir os vetores
A = np.array([1, 2])
B = np.array([3, 4])

# Calcular o produto escalar
produto_escalar = np.dot(A, B)

# Calcular os módulos dos vetores
modulo_A = np.linalg.norm(A)
modulo_B = np.linalg.norm(B)

# Calcular o cosseno do ângulo entre os vetores
cos_theta = produto_escalar / (modulo_A * modulo_B)
print("Cosseno: ", cos_theta)

# Calcular o ângulo em radianos
theta_rad = np.arccos(cos_theta)

# Converter o ângulo para graus
theta_deg = np.degrees(theta_rad)
print("Ângulo: ", theta_deg)
```

### Exercício: Ângulo $θ$ de dois vetores
Temos os vtores $u=(2,2)$ e $v=(0-2)$ encontre o ângulo θ entre eles.<br>
___
Resolução:<br>
$$ \cos(\theta) = \frac{\vec{A} \cdot \vec{B}}{\|\vec{A}\| \|\vec{B}\|} $$
<br><br>
$\vec{u} \cdot \vec{v} = (2,2) \cdot (0,-2) = -4$
<br>
$\|\vec{u}\| = \sqrt{2^2 + 2^2} = \sqrt{8}$
<br>
$\|\vec{v}\| = \sqrt{0^2 + (-2)^2} = \sqrt{4} = 2$
<br>
<br>

## Paralelismo de dois vetores

Dois vetores $\vec{A}$ e $\vec{B}$ são considerados paralelos se um for um múltiplo escalar do outro. Em outras palavras, se $\vec{A}$ for um múltiplo escalar de $\vec{B}$ (ou vice-versa), então os vetores são paralelos.

Matematicamente, isso pode ser expresso como:

$$ \vec{A} = k \cdot \vec{B} $$

ou

$$ \vec{B} = k \cdot \vec{A} $$

onde $k$ é um escalar.

<br><br>
Tambem podemos fazer a seguinte equação
$$\frac{x_1}{x_2} = \frac{y_1}{y_2} = a$$
Que com ela é gerado um escalar

Em Python, usando NumPy, você pode verificar o paralelismo de dois vetores facilmente. Aqui está um exemplo:

```python
# Definir os vetores
A = np.array([1, 2])
B = np.array([2, 4])
k = 2

# Verificar o paralelismo
if np.array_equal(k * A, B) or np.array_equal(k * B, A):
    print("Os vetores são paralelos.")
else:
    print("Os vetores não são paralelos.")
```

## Vetores Ortogonais
Dois vetores $\vec{A}$ e $\vec{B}$ são ortogonais se o ângulo entre eles é de $90^\circ$ (ou $\frac{\pi}{2}$ radianos). Isso significa que o produto escalar entre os vetores é zero:

$$ \vec{A} \cdot \vec{B} = 0 $$

Em outras palavras, se os vetores $\vec{A}$ e $\vec{B}$ forem ortogonais, isso significa que eles não têm nenhuma componente em comum na mesma direção.

Em Python, usando NumPy, você pode verificar se dois vetores são ortogonais facilmente. Aqui está um exemplo:


```python
# Definir os vetores
A = np.array([1, 0])
B = np.array([0, 1])

# Verificar se os vetores são ortogonais
if np.dot(A, B) == 0:
    print("Os vetores são ortogonais.")
else:
    print("Os vetores não são ortogonais.")
```

## Vetores Perpendiculares

Dois vetores $\vec{A}$ e $\vec{B}$ são perpendiculares se o ângulo entre eles é de $90^\circ$ (ou $\frac{\pi}{2}$ radianos). Isso significa que o produto escalar entre os vetores é zero:

$$ \vec{A} \cdot \vec{B} = 0 $$

Em outras palavras, se os vetores $\vec{A}$ e $\vec{B}$ forem perpendiculares, isso significa que eles formam um ângulo reto entre si.

## Projeção Ortogonal de um Vetor

A projeção ortogonal de um vetor $\vec{V}$ em relação a outro vetor $\vec{U}$ é a componente de $\vec{V}$ que está na direção de $\vec{U}$. Matematicamente, a projeção ortogonal de $\vec{V}$ em relação a $\vec{U}$ é dada por:

$$ \text{proj}_{\vec{U}}(\vec{V}) = \frac{\vec{V} \cdot \vec{U}}{\|\vec{U}\|^2} \cdot \vec{U} $$

Em outras palavras, a projeção ortogonal de $\vec{V}$ em relação a $\vec{U}$ é a componente de $\vec{V}$ na direção de $\vec{U}$, multiplicada pelo comprimento de $\vec{U}$.

Em Python, usando NumPy, você pode calcular a projeção ortogonal de um vetor $\vec{V}$ em relação a outro vetor $\vec{U}$ facilmente. Aqui está um exemplo:

```python
# Definir os vetores
V = np.array([3, 4])
U = np.array([1, 0])

# Calcular a projeção ortogonal de V em relação a U
proj_U_V = np.dot(V, U) / np.dot(U, U) * U
```

Uma combinação linear de vetores é uma expressão na forma de uma soma ponderada dos vetores, onde os coeficientes das ponderações são escalares.

Se tivermos $n$ vetores $\vec{v}_1, \vec{v}_2, ..., \vec{v}_n$ e escalares $c_1, c_2, ..., c_n$, a combinação linear desses vetores é dada por:

$$ c_1 \cdot \vec{v}_1 + c_2 \cdot \vec{v}_2 + ... + c_n \cdot \vec{v}_n $$

Os vetores e os escalares podem pertencer a qualquer espaço vetorial.

Em Python, usando NumPy, você pode calcular a combinação linear de vetores facilmente. Aqui está um exemplo:


```python
# Definir os vetores
v1 = np.array([1, 2])
v2 = np.array([3, 4])

# Definir os escalares
c1 = 2
c2 = -1

# Calcular a combinação linear dos vetores
comb_linear = c1 * v1 + c2 * v2

# Resultado: comb_linear = [1, 0]
```
